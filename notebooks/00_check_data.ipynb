# 00 — Check StatsBomb Data Availability

Run this first to see exactly which competitions and seasons are available.
Note down the `competition_id` and `season_id` you want to use.

```python
from statsbombpy import sb
import pandas as pd
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 20)
```

```python
# Load all available competitions
competitions = sb.competitions()
print(f'Total competitions available: {len(competitions)}')
competitions[['competition_name', 'season_name', 'competition_id', 'season_id', 'competition_gender']].sort_values(['competition_name', 'season_name'])
```

```python
# Filter to men's competitions only
mens = competitions[competitions['competition_gender'] == 'male']
print(f'Men competitions: {len(mens)}')
mens[['competition_name', 'season_name', 'competition_id', 'season_id']]
```

```python
# Count matches per competition to see which are most data-rich
for _, row in mens.iterrows():
    try:
        matches = sb.matches(competition_id=row['competition_id'], season_id=row['season_id'])
        print(f"{row['competition_name']} {row['season_name']}: {len(matches)} matches (comp_id={row['competition_id']}, season_id={row['season_id']})")
    except:
        print(f"{row['competition_name']} {row['season_name']}: ERROR")
```
